```python
help(cv2.kmeans)
```

```python
print(img.shape)
vimg=img.reshape((-1,3))
```

```python
vimg=np.float32(vimg)
vimg.shape
```

```python
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
```

```python
k=2
attempts=20
ret,lable,centre=cv2.kmeans(vimg,k,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
```

```python
centre=np.uint8(centre)
res=centre[lable.flatten()]
result_image=res.reshape((img.shape))
```

```python
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(result_image)
plt.show()
```
